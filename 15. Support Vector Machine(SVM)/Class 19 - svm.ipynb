```python
df = pd.read_csv('circle.csv')
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.head()
```

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter(df['feature1'], df['feature2'])
```

```python
df.label_y.value_counts()
```

```python
df1 = df[df.label_y == 0]
df2 = df[df.label_y == 1]

plt.scatter(df1.feature1 , df1.feature2, color='red', label='Class 0')
plt.scatter(df2.feature1 , df2.feature2, color='black', label='Class 1')

plt.legend()
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df)
```

```python
train.shape
```

```python
test.shape
```

```python
df.shape
```

# SVC

```python
from sklearn.svm import SVC # FRO REGRESSOR SVR
```

```python
svm1 = SVC(kernel = 'linear')
```

```python
svm1.fit(train[['feature1', 'feature2']], train['label_y'])
```

```python
svm1.score(test[['feature1', 'feature2']], test['label_y'])
```

# Poly

```python
df['f1_sq'] = df['feature1'] ** 2
df['f2_sq'] = df['feature2'] ** 2
df['f1f2'] = df['feature1'] * df['feature2']
```

```python
df.head()
```

```python
x = df.drop('label_y', axis=1)
```

```python
y = df['label_y']
```

```python
x.shape
```

```python
import plotly.express as px
px.scatter_3d(df, x='f1_sq', y='f2_sq', z='f1f2', color='label_y')
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=.30)
```

```python
xtrain.shape
```

```python
svm2 = SVC(kernel = 'linear')
```

```python
svm2.fit(xtrain, ytrain)
```

```python
xtrain.shape
```

```python
xtrain.head()
```

```python
svm2.score(xtest, ytest)
```
