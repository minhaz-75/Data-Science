# Linear Regression using Gradient Descent 

```python
m = 0
c = 0
l = 0.0001
n = int(len(x))
iteration = 1000

for i in range(iteration):
    pred = (m*x) + c
    dm = (-2/n)*sum(x*(y-pred))
    dc = (-2/n)*sum(y-pred)

    m = m-(l*dm)
    c = c-(l*dc)

print(m)
print(c)
```
